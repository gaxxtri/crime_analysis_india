```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)

DATA_PATH = "../data/processed/"
master = pd.read_csv(DATA_PATH + "MASTER_statewise_core_crime_dataset.csv")
```

```python
master.shape
master.head()
```

```python
master.info()
master.describe()
```

```python
master.sort_values("2022", ascending=False)[["state_ut", "2022"]].head(10)
master.sort_values("2022")[["state_ut", "2022"]].head(10)
```

IPC crimes are concentrated in a small number of highly populated states

```python
#Crime Rate Distribution
plt.figure(figsize=(8,4))
sns.histplot(master["Rate of Cognizable Crimes (IPC) (2022)"], bins=15, kde=True)
plt.title("Distribution of IPC Crime Rate (2022)")
plt.xlabel("Crime Rate per 100,000 Population")
plt.show()
```

Crime rates vary widely across states, indicating uneven crime burden

```python
#Literacy vs IPC Crime Rate
sns.scatterplot(
    data=master,
    x="literacy_rate_2011",
    y="Rate of Cognizable Crimes (IPC) (2022)"
)
plt.title("Literacy vs IPC Crime Rate")
plt.show()
```

```python
#Urbanization vs IPC Crime Rate
sns.scatterplot(
    data=master,
    x="urbanization_rate_2011",
    y="Rate of Cognizable Crimes (IPC) (2022)"
)
plt.title("Urbanization vs IPC Crime Rate")
plt.show()
```

```python
#Correlation Matrix
cols = [
    "Rate of Cognizable Crimes (IPC) (2022)",
    "literacy_rate_2011",
    "urbanization_rate_2011",
    "population_2011"
]

master[cols].corr()
```
