```python
#StateUT-wise Number of Women and Girls Victims of Rape during 2022
df_rape = pd.read_csv(
    DATA_PATH + "StateUT-wise Number of Women and Girls Victims of Rape during 2022.csv"
)
```

```python
df_rape.drop(columns=["Sl. No."], inplace=True)
df_rape.rename(columns={"State/UT": "state_ut"}, inplace=True)
df_rape["state_ut"] = df_rape["state_ut"].str.strip().str.title()
```

```python
df_rape.rename(
    columns={
        "Cases Reported - Col.(3)": "rape_cases_reported",
        "Child Victims of Rape (Below 18 Yrs) - Total Girl / Child Victims - Col.(8)": "child_rape_victims",
        "Women Victims of Rape (Above 18 Years) - Total Women / Adult Victims - Col.(13)": "adult_rape_victims",
        "Total Victims (Col.8+Col.13) - Col.(14)": "total_rape_victims"
    },
    inplace=True
)
```

```python
num_cols = df_rape.columns.difference(["state_ut"])
df_rape[num_cols] = df_rape[num_cols].apply(
    pd.to_numeric, errors="coerce"
)
```

```python
df_rape.isna().sum()
df_rape.fillna(0, inplace=True)
```

```python
df_rape["child_rape_share"] = (
    df_rape["child_rape_victims"] / df_rape["total_rape_victims"]
)

df_rape["adult_rape_share"] = (
    df_rape["adult_rape_victims"] / df_rape["total_rape_victims"]
)
```

```python
df_rape[
    ["child_rape_share", "adult_rape_share"]
].describe()
```

```python
df_rape.to_csv(
    "../data/processed/rape_victims_2022_cleaned.csv",
    index=False
)

```

```python
import os

OUT_PATH = "../data/processed/"
os.makedirs(OUT_PATH, exist_ok=True)
```

```python
df_convicts = pd.read_csv(DATA_PATH + "StateUT-wise Number of Convicts Confined in Major Indian Penal Code (IPC) Crimes from 2020 to 2022.csv")

df_convicts = (
    df_convicts
    .drop(columns=["Sl. No."])
    .rename(columns={"State/UT": "state_ut"})
)

df_convicts["state_ut"] = df_convicts["state_ut"].str.strip().str.title()

num_cols = df_convicts.columns.difference(["state_ut"])
df_convicts[num_cols] = df_convicts[num_cols].apply(pd.to_numeric, errors="coerce")

df_convicts.fillna(0, inplace=True)

df_convicts.to_csv(OUT_PATH + "ipc_convicts_cleaned.csv", index=False)
```

```python
df_kidnap = pd.read_csv(DATA_PATH + "StateUT-wise Number of Recovery of Kidnapped and Abducted Persons during 2022.csv")

df_kidnap = (
    df_kidnap
    .drop(columns=["Sl. No."])
    .rename(columns={"State/UT": "state_ut"})
)

df_kidnap["state_ut"] = df_kidnap["state_ut"].str.strip().str.title()

num_cols = df_kidnap.columns.difference(["state_ut"])
df_kidnap[num_cols] = df_kidnap[num_cols].apply(pd.to_numeric, errors="coerce")

df_kidnap.fillna(0, inplace=True)

df_kidnap.to_csv(OUT_PATH + "kidnapping_recovery_cleaned.csv", index=False)
```

```python
df_motive = pd.read_csv(DATA_PATH + "StateUT-wise Number of Motives of Murder during 2022.csv")

df_motive = (
    df_motive
    .drop(columns=["Sl. No."])
    .rename(columns={"State/UT": "state_ut"})
)

df_motive["state_ut"] = df_motive["state_ut"].str.strip().str.title()

num_cols = df_motive.columns.difference(["state_ut"])
df_motive[num_cols] = df_motive[num_cols].apply(pd.to_numeric, errors="coerce")

df_motive.fillna(0, inplace=True)

df_motive.to_csv(OUT_PATH + "murder_motives_cleaned.csv", index=False)
```

```python
df_purpose = pd.read_csv(
    DATA_PATH + "Category-wise Number of Purpose of Kidnapping and Abduction (Alive Victims) during 2022.csv"
)

df_purpose = df_purpose.drop(columns=["Sl. No."])

df_purpose.columns = (
    df_purpose.columns.str.lower()
    .str.replace(r'col\.\(\d+\)', '', regex=True)
    .str.replace('[^a-z0-9_]', '_', regex=True)
)

num_cols = df_purpose.columns.difference(["purpose"])
df_purpose[num_cols] = df_purpose[num_cols].apply(pd.to_numeric, errors="coerce")

df_purpose.fillna(0, inplace=True)

df_purpose.to_csv(
    OUT_PATH + "kidnapping_purpose_cleaned.csv",
    index=False
)
```

```python
df_rec_cat = pd.read_csv(
    DATA_PATH + "Category-wise Number of Recovery of Kidnapped and Abducted Persons during 2022.csv"
)

df_rec_cat = df_rec_cat.drop(columns=["Sl. No."])

df_rec_cat.columns = (
    df_rec_cat.columns.str.lower()
    .str.replace(r'col\.\(\d+\)', '', regex=True)
    .str.replace('[^a-z0-9_]', '_', regex=True)
)

df_rec_cat = df_rec_cat.rename(columns={"category": "recovery_category"})

num_cols = df_rec_cat.columns.difference(["recovery_category", "recovery"])
df_rec_cat[num_cols] = df_rec_cat[num_cols].apply(pd.to_numeric, errors="coerce")

df_rec_cat.fillna(0, inplace=True)

df_rec_cat.to_csv(
    OUT_PATH + "kidnapping_recovery_category_cleaned.csv",
    index=False
)
```

```python
df_child = pd.read_csv(
    DATA_PATH + "StateUT-wise Number of Child Victims of Kidnapping and Abduction during 2022.csv"
)

df_child = (
    df_child
    .drop(columns=["Sl. No."])
    .rename(columns={"State/UT": "state_ut"})
)

df_child["state_ut"] = df_child["state_ut"].str.strip().str.title()

num_cols = df_child.columns.difference(["state_ut"])
df_child[num_cols] = df_child[num_cols].apply(pd.to_numeric, errors="coerce")

df_child.fillna(0, inplace=True)

df_child.to_csv(
    OUT_PATH + "child_kidnapping_cleaned.csv",
    index=False
)
```

```python
df_foreign = pd.read_csv(
    DATA_PATH + "StateUTs -wise Number of Distribution of Foreign Undertrials in Jails by Gender and Age-groups.csv"
)

df_foreign = (
    df_foreign
    .drop(columns=["Sl. No."])
    .rename(columns={"State/UT": "state_ut"})
)

df_foreign["state_ut"] = df_foreign["state_ut"].str.strip().str.title()

num_cols = df_foreign.columns.difference(["state_ut"])
df_foreign[num_cols] = df_foreign[num_cols].apply(pd.to_numeric, errors="coerce")

df_foreign.fillna(0, inplace=True)

df_foreign.to_csv(
    OUT_PATH + "foreign_undertrials_cleaned.csv",
    index=False
)
```

```python
import pandas as pd

P = "../data/processed/"

ipc = pd.read_csv(P + "ipc_crimes_cleaned.csv")
pop = pd.read_csv(P + "population_literacy_urbanization_cleaned.csv")
murder = pd.read_csv(P + "murder_cases_2020_2022_cleaned.csv")
rape = pd.read_csv(P + "rape_victims_2022_cleaned.csv")
convicts = pd.read_csv(P + "ipc_convicts_cleaned.csv")
kidnap_state = pd.read_csv(P + "kidnapping_recovery_cleaned.csv")
motives = pd.read_csv(P + "murder_motives_cleaned.csv")
child_kidnap = pd.read_csv(P + "child_kidnapping_cleaned.csv")

MASTER = (
    ipc
    .merge(pop, on="state_ut", how="left")
    .merge(murder, on="state_ut", how="left", suffixes=("", "_murder"))
    .merge(rape, on="state_ut", how="left")
    .merge(convicts, on="state_ut", how="left", suffixes=("", "_convicts"))
    .merge(kidnap_state, on="state_ut", how="left", suffixes=("", "_kidnap"))
    .merge(motives, on="state_ut", how="left", suffixes=("", "_motive"))
    .merge(child_kidnap, on="state_ut", how="left", suffixes=("", "_child"))
)

MASTER.fillna(0, inplace=True)

MASTER.to_csv(
    P + "MASTER_statewise_core_crime_dataset.csv",
    index=False
)

MASTER.shape
```

```python
purpose = pd.read_csv(P + "kidnapping_purpose_cleaned.csv")
recovery_cat = pd.read_csv(P + "kidnapping_recovery_category_cleaned.csv")

purpose["source"] = "purpose_wise"
recovery_cat["source"] = "recovery_category_wise"

KIDNAPPING_CATEGORY = pd.concat(
    [purpose, recovery_cat],
    ignore_index=True
)

KIDNAPPING_CATEGORY.to_csv(
    P + "KIDNAPPING_categorywise_dataset.csv",
    index=False
)
```

```python
sll = pd.read_csv(P + "sll_crimes_cleaned.csv")
foreign = pd.read_csv(P + "foreign_undertrials_cleaned.csv")

SLL_FOREIGN = sll.merge(
    foreign,
    on="state_ut",
    how="left",
    suffixes=("_sll", "_foreign")
)

SLL_FOREIGN.fillna(0, inplace=True)

SLL_FOREIGN.to_csv(
    P + "SLL_and_ForeignUndertrials_dataset.csv",
    index=False
)
```
